## 4. Construcción de la variable objetivo

A partir de la columna **Texto causa**, se construye una variable objetivo multiclase llamada **Modo de fallo**, agrupada en 8 categorías principales.

```python
# ==========================================
# 🏷 CONSTRUCCIÓN DE LA VARIABLE OBJETIVO
# ==========================================

clases_interes = [
    "Mecánico - eléctrico general",
    "Neumático - valvula/Aire",
    "Mecánico - Fugas/Lubricación",
    "Eléctrico - Potencia/Aislamiento",
    "Eléctrico/Control - PLC/Sensores/Conexiones",
    "Intervención General (Cambio/Ajuste)",
    "Mecánico (Ajuste/Cambio/Desgaste)",
    "Otros / Sin Clasificar"
]

diccionario_causas = {
    "Mecánico - eléctrico general": 0,
    "Neumático - valvula/Aire": 1,
    "Mecánico - Fugas/Lubricación": 2,
    "Eléctrico - Potencia/Aislamiento": 3,
    "Eléctrico/Control - PLC/Sensores/Conexiones": 4,
    "Intervención General (Cambio/Ajuste)": 5,
    "Mecánico (Ajuste/Cambio/Desgaste)": 6,
    "Otros / Sin Clasificar": 7
}

df_modelo = df[df["Texto causa"].isin(clases_interes)].copy()
df_modelo["objetivo_multiple"] = df_modelo["Texto causa"].map(diccionario_causas)

print("Distribución original de 'Texto causa':")
display(df["Texto causa"].value_counts(dropna=False))

print("\nDistribución de la variable objetivo multiclase:")
display(df_modelo["objetivo_multiple"].value_counts().sort_index())
```

```python
# ==========================================
# 📊 DISTRIBUCIÓN DE CLASES
# ==========================================

conteo_clases = df_modelo["objetivo_multiple"].value_counts().sort_index()

etiquetas_clases = {
    0: "Mecánico-Eléctrico",
    1: "Neumático",
    2: "Fugas/Lubricación",
    3: "Potencia/Aislamiento",
    4: "PLC/Sensores",
    5: "Intervención General",
    6: "Ajuste/Desgaste",
    7: "Otros"
}

plt.figure(figsize=(10, 5))
plt.bar([etiquetas_clases[i] for i in conteo_clases.index], conteo_clases.values)
plt.title("Distribución de la variable objetivo multiclase")
plt.xlabel("Clase")
plt.ylabel("Cantidad de registros")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## 5. Ingeniería de variables y preparación del dataset

Se transforman las columnas de fecha en variables derivadas y se seleccionan las variables de entrada y salida para el modelo.

```python
# ==========================================
# 🧹 INGENIERÍA DE VARIABLES
# ==========================================

datos = df_modelo.copy()

columnas_fecha = ["Inicio avería", "Fin de avería", "Fecha de cierre", "Fecha de aviso"]

for columna in columnas_fecha:
    if columna in datos.columns:
        datos[columna] = pd.to_datetime(datos[columna], errors="coerce")
        datos[f"{columna}_anio"] = datos[columna].dt.year
        datos[f"{columna}_mes"] = datos[columna].dt.month
        datos[f"{columna}_dia"] = datos[columna].dt.day
        datos[f"{columna}_dia_semana"] = datos[columna].dt.dayofweek

columnas_descartar = [
    "Clase de aviso",
    "objetivo_multiple",
    "Notificación",
    "Orden",
    "Inicio avería",
    "Fin de avería",
    "Fecha de cierre",
    "Fecha de aviso",
    "Texto causa"
]

X = datos.drop(columns=columnas_descartar, errors="ignore")
y = datos["objetivo_multiple"]

print("Dimensión de X:", X.shape)
print("Dimensión de y:", y.shape)
display(X.head())
```

```python
# ==========================================
# 🔢 SEPARACIÓN DE VARIABLES NUMÉRICAS Y CATEGÓRICAS
# ==========================================

columnas_numericas = X.select_dtypes(include=["number"]).columns.tolist()
columnas_categoricas = X.select_dtypes(include=["object", "category"]).columns.tolist()

print("Cantidad de variables numéricas:", len(columnas_numericas))
print("Cantidad de variables categóricas:", len(columnas_categoricas))

print("\nEjemplo de variables numéricas:")
print(columnas_numericas[:10])

print("\nEjemplo de variables categóricas:")
print(columnas_categoricas[:10])
```

```python
# ==========================================
# ✂️ DIVISIÓN ENTRE ENTRENAMIENTO Y PRUEBA
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Tamaño de X_train:", X_train.shape)
print("Tamaño de X_test:", X_test.shape)

print("\nDistribución de clases en entrenamiento:")
print(y_train.value_counts(normalize=True).sort_index())

print("\nDistribución de clases en prueba:")
print(y_test.value_counts(normalize=True).sort_index())
```

## 6. Construcción del pipeline y entrenamiento del modelo

Se construye un pipeline que incluye imputación de faltantes, codificación de variables categóricas y entrenamiento con Random Forest.

```python
# ==========================================
# 🤖 PIPELINE DE PREPROCESAMIENTO Y RANDOM FOREST
# ==========================================

transformador_numerico = Pipeline(steps=[
    ("imputador", SimpleImputer(strategy="median"))
])

transformador_categorico = Pipeline(steps=[
    ("imputador", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocesador = ColumnTransformer(
    transformers=[
        ("numericas", transformador_numerico, columnas_numericas),
        ("categoricas", transformador_categorico, columnas_categoricas)
    ]
)

modelo_rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    class_weight="balanced",
    n_jobs=-1
)

pipeline_rf = Pipeline(steps=[
    ("preprocesador", preprocesador),
    ("modelo", modelo_rf)
])

pipeline_rf.fit(X_train, y_train)
print("Entrenamiento completado exitosamente.")
```

## 7. Evaluación del modelo

Se calculan métricas globales, el reporte de clasificación, la matriz de confusión y la distribución de probabilidades predichas.

```python
# ==========================================
# 📌 PREDICCIONES DEL MODELO
# ==========================================

y_pred = pipeline_rf.predict(X_test)
y_prob = pipeline_rf.predict_proba(X_test)

print("Primeras 10 predicciones:", y_pred[:10])
print("Shape de predict_proba:", y_prob.shape)
```

```python
# ==========================================
# 📊 MÉTRICAS DE EVALUACIÓN
# ==========================================

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="weighted", zero_division=0)
recall = recall_score(y_test, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)

target_names = [
    "Mecánico-Eléctrico",
    "Neumático",
    "Fugas/Lubricación",
    "Potencia/Aislamiento",
    "PLC/Sensores",
    "Intervención General",
    "Ajuste/Desgaste",
    "Otros"
]

print("Accuracy :", round(accuracy, 4))
print("Precision (Weighted):", round(precision, 4))
print("Recall    (Weighted):", round(recall, 4))
print("F1-score  (Weighted):", round(f1, 4))

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, target_names=target_names, zero_division=0))
```

```python
# ==========================================
# 📉 MATRIZ DE CONFUSIÓN
# ==========================================

matriz_confusion = confusion_matrix(y_test, y_pred)

print("Matriz de confusión:")
print(matriz_confusion)

fig, ax = plt.subplots(figsize=(8, 8))
disp = ConfusionMatrixDisplay(
    confusion_matrix=matriz_confusion,
    display_labels=target_names
)
disp.plot(ax=ax, xticks_rotation=45, cmap="Blues", colorbar=False)
plt.title("Matriz de confusión - Random Forest")
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# 📌 IMPORTANCIA DE VARIABLES
# ==========================================

nombres_variables = pipeline_rf.named_steps["preprocesador"].get_feature_names_out()
importancias = pipeline_rf.named_steps["modelo"].feature_importances_

df_importancias = pd.DataFrame({
    "variable": nombres_variables,
    "importancia": importancias
}).sort_values(by="importancia", ascending=False)

top_15 = df_importancias.head(15)
display(top_15)

plt.figure(figsize=(10, 7))
plt.barh(top_15["variable"][::-1], top_15["importancia"][::-1])
plt.title("Top 15 variables más importantes")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.show()
```

```python
# ==========================================
# 📈 DISTRIBUCIÓN DE PROBABILIDADES PREDICHAS
# ==========================================

clase_idx = 1
nombre_clase = target_names[clase_idx]

plt.figure(figsize=(8, 5))
plt.hist(y_prob[:, clase_idx], bins=20)
plt.title(f"Distribución de probabilidades predichas para la clase {nombre_clase}")
plt.xlabel(f"Probabilidad predicha de {nombre_clase}")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# ==========================================
# 🧠 CONCLUSIONES FINALES DEL MODELO
# ==========================================

tasa_error = 1 - accuracy

print("=" * 80)
print("CONCLUSIONES FINALES DEL MODELO")
print("=" * 80)
print(f"1. El modelo Random Forest alcanzó una exactitud de {accuracy:.2%}.")
print(f"2. La precisión ponderada fue de {precision:.2%}.")
print(f"3. El recall ponderado fue de {recall:.2%}.")
print(f"4. El F1-score ponderado fue de {f1:.2%}.")
print(f"5. La tasa de error global fue de {tasa_error:.2%}.")
print("6. La matriz de confusión corresponde a un problema multiclase de 8 categorías, no binario.")
print("7. Las clases con menos registros suelen presentar más errores de clasificación.")
print("8. La revisión de importancias de variables permite identificar qué factores influyen más en la predicción.")
```

## 8. Predicción de fallas a 30 días con Prophet

En esta sección se utiliza Prophet para proyectar el comportamiento esperado de los modos de fallo en los próximos 30 días.

```python
# ==========================================
# 🔮 INSTALACIÓN E IMPORTACIÓN DE PROPHET
# ==========================================

!pip install prophet -q

from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
# ==========================================
# 📂 CARGA DE DATOS PARA FORECAST
# ==========================================

df_forecast = pd.read_excel(archivo_excel, sheet_name="IW69 AVISOS ENG")

print("Dimensiones:", df_forecast.shape)
display(df_forecast.head())
```

```python
# ==========================================
# 🧹 LIMPIEZA Y CONSTRUCCIÓN DE MODO DE FALLO
# ==========================================

df_forecast["Fecha de aviso"] = pd.to_datetime(df_forecast["Fecha de aviso"], errors="coerce")
df_forecast = df_forecast.dropna(subset=["Fecha de aviso", "Texto causa"]).copy()

mapeo_modo_fallo = {
    "Mecánico - eléctrico general": "Mecánico-Eléctrico",
    "Neumático - valvula/Aire": "Neumático",
    "Mecánico - Fugas/Lubricación": "Fugas/Lubricación",
    "Eléctrico - Potencia/Aislamiento": "Potencia/Aislamiento",
    "Eléctrico/Control - PLC/Sensores/Conexiones": "PLC/Sensores",
    "Intervención General (Cambio/Ajuste)": "Intervención General",
    "Mecánico (Ajuste/Cambio/Desgaste)": "Ajuste/Desgaste",
    "Otros / Sin Clasificar": "Otros"
}

df_forecast["Modo de fallo"] = df_forecast["Texto causa"].map(mapeo_modo_fallo)
df_forecast = df_forecast.dropna(subset=["Modo de fallo"]).copy()

print("Distribución de Modo de fallo:")
display(df_forecast["Modo de fallo"].value_counts())
```

```python
# ==========================================
# 📅 CONSTRUCCIÓN DE SERIE TEMPORAL
# ==========================================

serie_diaria = (
    df_forecast
    .groupby([df_forecast["Fecha de aviso"].dt.date, "Modo de fallo"])
    .size()
    .reset_index(name="cantidad")
)

serie_diaria = serie_diaria.rename(columns={"Fecha de aviso": "fecha"})
serie_diaria["fecha"] = pd.to_datetime(serie_diaria["fecha"])

print("Vista previa de la serie diaria:")
display(serie_diaria.head())
```

```python
# ==========================================
# 🎯 SELECCIÓN DE UNA CATEGORÍA PARA FORECAST
# ==========================================

conteo_por_categoria = serie_diaria["Modo de fallo"].value_counts()
display(conteo_por_categoria)

categoria = "Neumático"
df_cat = serie_diaria[serie_diaria["Modo de fallo"] == categoria].copy()

print("Registros para la categoría seleccionada:", len(df_cat))
display(df_cat.head())
```

```python
# ==========================================
# 🔄 FORMATO REQUERIDO POR PROPHET
# ==========================================

df_prophet = df_cat.rename(columns={
    "fecha": "ds",
    "cantidad": "y"
})[["ds", "y"]].sort_values("ds")

rango_fechas = pd.DataFrame({
    "ds": pd.date_range(start=df_prophet["ds"].min(), end=df_prophet["ds"].max(), freq="D")
})

df_prophet = rango_fechas.merge(df_prophet, on="ds", how="left")
df_prophet["y"] = df_prophet["y"].fillna(0)

display(df_prophet.head())
display(df_prophet.tail())
```

```python
# ==========================================
# 🤖 ENTRENAMIENTO DE PROPHET
# ==========================================

modelo_prophet = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False
)

modelo_prophet.fit(df_prophet)
```
